```python
menu = {
    "popcorn": 250,
    "nachos": 350,
    "burger": 280,
    "hotdog": 300,
    "fries": 220,
    "samosa": 100,
    "sandwich": 200,
    "pizza": 400,
    "cola": 200,
    "water": 50,
    "coffee": 180,
    "tea": 120,
    "candy": 150,
    "brownie": 220,
    "icecream": 180
}

cart = {}
total = 0

print("------------MENU------------")
for key, value in menu.items():
    print(f"{key:10}: ₹{value:3}")
print("----------------------------")

while True:
    food = input("What would you like to order (q to quit): ").lower()
    
    if food == "q":
        break
    elif menu.get(food) is not None:
        try:
            qty = int(input(f"Enter quantity for {food}: "))
            
            # Validate quantity range before adding to cart
            if qty <= 0:
                print("Please enter a number greater than 0.")
            elif qty > 20:
                print("Quantity for a single item cannot exceed 20.")
            else:
                cart[food] = cart.get(food, 0) + qty
                print(f"Added {qty} {food}(s) to your cart.")
                
        except ValueError:
            print("Please enter a valid number.")
    else:
        print("Sorry, that item is not on the menu.")

if not cart:
    print("\nYour cart is empty. Thanks for visiting!")
else:
    print("\n-------YOUR ORDER-------")
    for food, qty in cart.items():
        item_total = menu.get(food) * qty
        total += item_total
        print(f"{food:10} x {qty}: ₹{item_total}")
    print("------------------------")
    print(f"THE TOTAL BILL IS ₹{total}")
```
