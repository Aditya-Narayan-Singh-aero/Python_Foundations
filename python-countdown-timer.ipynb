```python
import time

timer  = int(input("Enter the number of seconds for timer: "))
if timer < 0:
    print("Time cannot be less than 0")
elif True:
    while timer > 0:
        seconds = timer%60
        minutes = int(timer/60)%60
        hours = int(timer/3600)%24
        print(f"{hours:02d}::{minutes:02d}::{seconds:02d}")
        time.sleep(1)
        timer-=1
    print("Times Up")
```
