```python
questions = ("Which organelle is known as the 'powerhouse' of the cell?",
            "What is the chemical symbol for gold??",
            "Which planet in our solar system is commonly known as the Red Planet?",
            "What is the SI unit of electric current?",
           "What is the process called when a gas turns into a liquid?")

options = (("a) Nucleus", "b) Ribosome", "c) Mitochondria", "d) Chloroplast"),
    ("a) Au", "b) Ag", "c) Fe", "d) Cu"),
    ("a) Venus", "b) Mars", "c) Jupiter", "d) Saturn"),
    ("a) Volt", "b) Watt", "c) Ampere", "d) Joule"),
    ("a) Condensation", "b) Evaporation", "c) Sublimation", "d) Melting"))

answers = ("c","a","b","c","a")

guesses = []

score = 0

question_number = 0

for question in questions:
    print("--------------------------")
    print(question)
    for option in options[question_number]:
        print(option)
    while True:
        guess = input("Enter(A,B,C,D): ").lower()
        if guess in ("a", "b", "c", "d"):
            break
        else:   
            print("Incorrect option!!")
            print("Try again")
        
    guesses.append(guess)
    if guess == answers[question_number]:
        score += 1
        print("CORRECT!!")
    else:
        print("INCORRECT!!")
        print(f"correct answer is {answers[question_number]}")
    question_number += 1

print("--------------------------")
print("          RESULTS         ")
print("--------------------------")

print("answers: ", end="")
for answer in answers:
    print(answer, end=" ")
print()
print("guesses: ", end="")
for guess in guesses:
    print(guess, end=" ")
print()
print("--------------------------")
print(f"your final score is {score} / 5")
```
