```python
items = []
prices = []

while True:
    item = input("Enter an item to buy(q to quit): ")
    if item.lower() == "q":
        break
    else:
        quantity = int(input(f"how many {item}(s) do you want?"))
        unit_price = float(input(f"Enter the price of a {item} ₹"))
        total_price = quantity * unit_price
        items.append(f"{item}(x{quantity})")
        prices.append(total_price)
print("-----YOUR CART-----")
for item,price in zip(items,prices):
    print(f"{item}---{price}₹")
    
total = sum(prices)
print("---------------------")
print(f"Your total is ₹{total}")    
       
```
